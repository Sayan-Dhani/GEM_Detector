```python
import ROOT
from array import array

# Assuming x and y are your data points
x = t_p_new
y = g_new  # Replace this with your actual data

# Create a TGraph
co_gr = ROOT.TGraph(len(x), array('d', x), array('d', y))

# Define your custom fit function
def custom_fit_function(x, params):
    a, b = params[0], params[1]
    return a * ROOT.TMath.Exp(b * x[0])

# Create a TF1 function with the custom fit function
fit_function_root_gain = ROOT.TF1("fit_function_root_gain", custom_fit_function, min(x), max(x), 2)

# Set initial parameter values
initial_a_value =  0.5
initial_b_value = 0.03
fit_function_root_gain.SetParameter(0, initial_a_value)
fit_function_root_gain.SetParameter(1, initial_b_value)

# Fit the TGraph using the fit function
fit_result_gain = co_gr.Fit(fit_function_root_gain, "S")

# # Print the fit result
# fit_result_gain.Print()

# Optionally, get the fitted parameter values
fitted_a = fit_function_root_gain.GetParameter(0)
fitted_b = fit_function_root_gain.GetParameter(1)

# Optionally, get the errors on the fitted parameters
error_a = fit_function_root_gain.GetParError(0)
error_b = fit_function_root_gain.GetParError(1)
chi2 = "{:.2e}".format(fit_result_gain.Chi2())
chi2_value = float(chi2)
NDf = fit_result_gain.Ndf()


c5 = root.TCanvas("","",1000,600)

co_gr.SetTitle("; T/p (K/atm); gain")
co_gr.GetXaxis().SetTitleSize(0.05)
co_gr.GetXaxis().SetTitleColor(1)
co_gr.GetYaxis().SetTitleSize(0.05)
co_gr.GetYaxis().SetTitleColor(1)
co_gr.GetYaxis().SetTitleOffset(0.9)

co_gr.GetYaxis().SetRangeUser(2000,10000)
# co_gr.GetXaxis().SetRangeUser(0.28,0.32)

co_gr.SetMarkerStyle(8)
co_gr.SetMarkerColor(1)
co_gr.SetMarkerSize(0.5)
co_gr.Draw("AP")

fit_function_root_gain.SetLineWidth(3)
fit_function_root_gain.Draw("SAME")

box = ROOT.TPaveText(0.15, 0.87, 0.45, 0.7, "NDC")
box.SetFillColor(0)
box.SetFillStyle(2)
box.SetLineColor(1)
box.SetTextSize(0.03)

latex = ROOT.TLatex()
latex.SetTextSize(0.03)
latex.SetTextAlign(13)
latex.DrawLatexNDC(0.15, 0.85, f"$$A: {fitted_a:.2f} \pm {error_a:.2f}$$")
latex.DrawLatexNDC(0.15, 0.80, f"$$B: {fitted_b:.2f} \pm {error_b:.2f}$$")
latex.DrawLatexNDC(0.15, 0.75, f"$$\chi^2/ndf: {chi2}/{NDf}$$")


print('fitted parameter "A" : ',np.round(fitted_a,4),'±',np.round(error_a,4))
print('fitted parameter "B" : ',np.round(fitted_b,4),'±',np.round(error_b,4))
print(chi2,NDf)

# latex.DrawLatexNDC(0.15, 0.70, f"NDf: {NDf}")

box.Draw()

c5.Draw()
c5.SaveAs("gain_fit.png")
c5.SaveAs("gain_fit.eps")
```
